```python
import numpy as np
import pandas as pd 

df = pd.read_csv(
    r"E:\merge_loan_refund_base.csv",
    sep=";",
    parse_dates=["loanDate","dueDate", "last_refund_date"]
)
print(df.shape)
df.head
```

```python
cols = [
    "clientID",
    "loanID",
    "loanDate",
    "dueDate",
    "last_refund_date",
    "payment_delta_days"
]

df = df[cols]
df.head()
```

```python
df_paid = df.dropna(subset=["payment_delta_days"])
print(df_paid.shape)
df_paid.head()
```

```python
df_paid = df_paid.sort_values(
    by=["clientID", "last_refund_date"],
    ascending=[True, False]
)

df_paid.head(10)
```

```python
df_paid["repay_rank"] = df_paid.groupby("clientID").cumcount() + 1
df_paid.head(10)
```

```python
last_loans = df_paid[df_paid["repay_rank"] <= 3].copy()
last_loans.head(10)
```

```python
loan_counts = last_loans.groupby("clientID")["repay_rank"].max()

def compute_weight(row):
    n = loan_counts[row["clientID"]]
    if n == 1:
        return 1.0
    elif n == 2:
        return 0.6 if row["repay_rank"] == 1 else 0.4
    else:  # n == 3
        return {1: 0.5, 2: 0.3, 3: 0.2}[row["repay_rank"]]

last_loans["weight"] = last_loans.apply(compute_weight, axis=1)
last_loans.head(10)
```

```python
last_loans["weighted_delta"] = (
    last_loans["payment_delta_days"] * last_loans["weight"]
)

last_loans[[
    "clientID",
    "repay_rank",
    "payment_delta_days",
    "weight",
    "weighted_delta"
]].head(10)
```

```python
behavior_score = (
    last_loans
    .groupby("clientID")["weighted_delta"]
    .sum()
    .reset_index(name="behavior_score_raw")
)

behavior_score.head(10)
```

```python
behavior_score.to_csv(
    r"E:\behavior_score_raw.csv",
    index=False,
    sep=";"
)

print("Raw behavior score saved")
```
